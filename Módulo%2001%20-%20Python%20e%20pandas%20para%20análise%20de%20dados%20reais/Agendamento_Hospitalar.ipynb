```
#Desafio 2 - Fazer o mesmo gráfico para o mês mais atual

#criando uma cópia dos dados
df2 = df

# #Tentando gerar o gráfico
# ax2 = df2.plot(x='Unidade da Federação', y='2020/Ago', kind='bar', 
#                figsize=(10,7))
# ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
# plt.title("Total de gastos com procedimentos hospitalares do SUS" 
#           + "por unidade da federação em Agosto/2020")
# plt.show()

#Deu erro por que temos um tipo de dado inválido na coluna 2020/Ago
```

```
#Convertendo o campo 2020/Ago para numérico e deixando o que não for numérico como nan
df2['2020/Ago'] = pd.to_numeric(df2['2020/Ago'], errors='coerce')
df2.info
```

```
#criando um novo dataframe e substituindo os valores nan por 0.00
df3 = df2.fillna(value=0.00)

#ordenando o df por valor
df3 = df2.sort_values('2020/Ago', ascending=False)

df3.info
```

```
#Recriando o gráfico após as correções nos dados

df3['2020/Ago'] = df3['2020/Ago']/1000000
ax3 = df3.plot(x='Unidade da Federação', y='2020/Ago', kind='barh', 
               figsize=(15,10), color='red', legend=False)

ax3.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title("Total de gastos com procedimentos hospitalares do SUS " 
          + "por unidade da federação em Agosto/2020", size=15)

plt.ylabel("Unidades da Federação", size=15)
plt.xlabel("Gastos em Milhões por UF", size=15)
plt.show()
```

# Agora fazendo a análise para dois meses

```
dados = df3
dados[['2008/Ago','2008/Set']].head()
```

```
dados.mean()
```

```
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,'Unidade da Federação')
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head()
```

```
usaveis.loc['12 Acre']
```

```
usaveis.T.head()
```

```
usaveis.T.tail()
```

```
usaveis = usaveis.drop('Total', axis=1)
usaveis.T.tail()
```

```
ax_usaveis = usaveis.T.plot(figsize=(15,10))

ax_usaveis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.legend(loc='center left',bbox_to_anchor=(1.0,0.5), title='UF')
plt.title("Total de gastos com procedimentos hospitalares do SUS" 
          + "por unidade da federação ao longo dos anos", size=15)

plt.xlabel("Ano/Mês", size=15)
plt.ylabel("Gastos x Ano/Mês", size=15)

plt.show()


```
